# Implementación de calculo de funciones anidadas con hilos, es decir
### g(x) -> Hilo 1
### f(x) -> Hilo 2

```python
from threading import Thread, Lock
from queue import Queue
import random
```

# Creando las clases
### Para la clase Mutex

```python
class mutex:
    
    def __init__(self):
        self.lock = Lock()
        
    def bloquea(self):
        self.lock.acquire()
        
    def libera(self):
        self.lock.release()
```

### Para la clase g(x)

```python
class g(Thread):
    
    #Creando el constructor
    #Recibe el valor de la x par la función
    #Y nuestro mutex para administrar la variable
    
    def __init__(self, x, mutex):
        
        super().__init__()
        self.x = x
        self.mutex = mutex
        
    # Definimos la operacion que realizara g(x)
    # Donde g(x) = 3x² - 3x³ + 2x 
    def operacion(self, x):
        
        global resultado
        
        resultado = ( (3*x)**2 ) - ( (3*x)**3 ) + ( (2*x) ) 
        print("g(x) = 3x² - 3x³ + 2x")
        print("g(x) = ",(3*x)**2,"-", (3*x)**3,"+",2*x)
        print("El resultado es:",resultado)
    
    def run(self):
        
        self.mutex.bloquea()
        print("\n\nSoy la g(x)")
        print("La x recibida es = ",self.x)
        self.operacion(self.x)
        self.mutex.libera()
```

### Para la clase f(x)

```python


class f(Thread):
 
    
    def __init__(self, mutex):
        
        super().__init__()
        self.mutex = mutex
    
    # Definimos la operacion que realizara g(x)
    # Donde f(x) = (2x +3x²) / 3x²
        
    def operacion(self, x):
        
        global resultado
        
        resultado = ( ( 2*x ) + ( (3*x)**2 ) ) * ( ( 3*x)**2 )
        print("f(x) =( 2x + 3x²) / 3x²")
        print("f(x) = ",2*x,"+", (3*x)**2 ,"*",(3*x)**2)
        print("El resultado es:",resultado)
        resultado = -1
        
    def run(self):

        
        while True:
            
            if resultado != 0:
                
                self.mutex.bloquea()
                print("\n\nSoy la f(x)")
                print("La x a evaluar es: ",resultado)
                self.operacion(resultado)
                self.mutex.libera()
                break
            
```

### Ejecutor (main)

```python
mutex = mutex()

resultado = 0

datos=[]
#Asignando valores en el arreglo x

i = -100

while i <= 100:
    datos.append(i)
    i+=0.25
    
#Se ejecutarán mientras esté vacía
while datos:

    gx = g(datos.pop(),mutex)
    fx = f(mutex)

    gx.start()
    fx.start()

    gx.join()
    fx.join()
```
