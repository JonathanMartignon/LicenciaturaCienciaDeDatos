```python
import threading
import queue 
from time import sleep
import random as rd
```

```python


#Creando la clase productor
class Productor(threading.Thread):
    
    global Cola_datos
    global itemCount
    global BUFFER_SIZE
    
    def __init__(self):
        threading.Thread.__init__(self)
        self.wake = True
        self.item = ''
            
    def run(self):
        
        while self.wake:
            
            self.item = produceItem()

            if (itemCount == BUFFER_SIZE):
                self.wake = to_sleep()

            putItemIntoBuffer(self.item)

            if (itemCount == 1):
                wakeupC()
                
```
