```python
#Importamos las librerias.
import numpy as np
import multiprocessing as mp
import random as rd
import time as tm
```

```python

#Creación de la función que come
#Definimos como que será realizada por par de filósofos
#Es decir, siempre habrá parejas comiendo
def comer (time, id1, id2,tiempo):
    '''
    Función comer, recibe 
    - El arreglo de los tiempos que le toma comer a cada filósofo (Time)
    - El índice del proceso al que hay que restarle tiempo id1 = filósofo 1, id2 = filósofo 2
    - El tiempo que le toca comer ( Es decir, el que vamos a restar a tiempos)
    
    ejemplo:
    
    time [30,30,30,30,30] Suponiendo que cada uno tardará 30 minutos en terminar de comer
    id1 = 0, id2 =2   Serán el turno del filósofo 1 y 3
    tiempo = 5 Será el tiempo actual que pueden comer
    al restar el tiempo actual que pueden comer al que les toma comer tenemos un arreglo:
    
    time [ 25, 30 , 25 , 30 , 30 ]
    
    '''
    
    #Verificaremos el arreglo compartido, donde cada parte representa el tiempo que le falta para terminar de comer (0)
    #Como trabajamos por parejas, la consideraremos como 1, es decir si uno de los 2 no ha  terminado de comer (0)
    #Continuará hasta que los 2 hayan terminado de comer (es decir, ambos tengan valores igual o menor a 0)
    if(time[id1] > 0 or time[id2] > 0):
        
        time[id1] -= tiempo
        time[id2] -= tiempo
        
        #Simulamos lo que los filósofos tardan en comer
        print("\nFilósofo ",id1+1," y filósofo ", id2+1, " tienen ", tiempo," minutos para comer")
        for i in range (tiempo):
            print(i+1)
            tm.sleep(1)
        print("Terminó turno de los filósofos ",id1+1," y ",id2+1,"\n\n")
        

Proc_1 =  rd.randint(0,4) #Proceso a ejecutar inicial, será aleatorio
Max_time = rd.randint(10,30) #Máximo de tiempo que se tardarán los filósofos en terminar de comer




Tiempos = mp.Array('i',[Max_time,Max_time,Max_time,Max_time,Max_time]) #Creación de un arreglo compartido que almacenará los datos de tiempo total de cada 'comida'


#Aquí podemos observar el tiempo que terminará cada filosofo su comida
#for i in range(0,5):
#    print("filosofo",i+1,"= ",Tiempos[i])


#Realizaremos el proceso siempre que tengamos un valor mayor a 0
# Esto quiere decir que existe algún filósofo que no ha terminado

while Tiempos [0] > 0 or Tiempos [1] > 0 or Tiempos [2] > 0 or Tiempos [3] > 0 or Tiempos [4] > 0:
#while Tiempos[i] > 0 for i in range(0,5)
#Intentamos verificar todos de una sola vez, no pudimos :c
  
  
  # Nos aseguramos de crear los turnos 1 y 2
  # Asegurandonos de que no sean adyacentes
  
    Proc_1 += 1
  
  
  
    if (Proc_1 < 4):

        if Proc_1 == 3:

            Proc_2 = 0

        else:
            Proc_2 = Proc_1 + 2
    else:
    
        if Proc_1 >= 5:

            Proc_1 = 0
            Proc_2 = 2

        else:
            
            Proc_2 = 1


    # Creamos el tiempo que van a tardar los filosofos actuales en comer su turno
    eat_time = rd.randint(1,4)

    Proceso1 = mp.Process(target= comer, args = (Tiempos,Proc_1,Proc_2,eat_time))


    Proceso1.start()
    Proceso1.join()
    
#Mostramos el arreglo final, los valores deben ser 0 o iguales a 0
print("Los tiempos finales quedaron: ")
for i in range (0,5):
    print("Filosofo ",i+1,Tiempos[i])
    
# Existen valores negativos, puesto que al tomarlos como parejas, puede ser que uno haya terminado
# Y su compañero aún no, es decir, puede que a algún filósofo haya terminado y a su compañero le falten
# 3 minutos.
```
