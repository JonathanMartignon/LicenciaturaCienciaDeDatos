# Instanciación de Threadings

```python
import threading
import time

#Construcción de una clase
class MyThread(threading.Thread):

    def __init__(self, number):
        #Heredamos las propiedades de Thread
        super(MyThread, self).__init__()
        self.number = number
    def run (self):
        time.sleep(0.1)
        print("Soy el hilo " + str(self.number))
thread_list = []
for i in range(5):
    thread = MyThread(i)
    thread_list.append(thread)
    thread.start()
```
